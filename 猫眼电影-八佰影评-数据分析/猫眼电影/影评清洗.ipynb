## 读取影评数据

```python
import pandas as pd
data=[]
with open('comments.txt', 'r',encoding='utf-8-sig') as f_input:
    for line in f_input:
        data.append(list(line.strip().split(',')))
data
```

## 转换为DataFrame

```python
df = pd.DataFrame(data).iloc[:, 0:6]
df.shape
```

## 添加列名

```python
df.columns = ['观众ID','观众昵称','城市','评论内容','评分','评论时间']
df.sample(5)
```

## 删除重复数据

```python
df = df.drop_duplicates()
print(df.shape)
```

```python
df.info()
df['评分'] = df['评分'].astype('float')
```
